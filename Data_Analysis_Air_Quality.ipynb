```python
def borough (dataset):
    if (("10" in dataset["Geo Join ID"]) and not ("20"in dataset["Geo Join ID"]) and not ("30"in dataset["Geo Join ID"]) and not ("40"in dataset["Geo Join ID"])):
        return "Bronx"
    elif ("20" in dataset["Geo Join ID"]):
        return "Brooklyn"
    elif ("30" in dataset["Geo Join ID"]):
        return "Manhattan"
    elif ("40"in dataset["Geo Join ID"]):
        return "Queens"
    elif ("50" in dataset["Geo Join ID"]):
        return "Staten Island"

df_same["Borough"] = df_same.apply(borough, axis=1)
```

```python
 df_same = df_same.groupby(["Borough"]).mean() 
```
